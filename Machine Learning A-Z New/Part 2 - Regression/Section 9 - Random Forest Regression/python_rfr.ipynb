```python
# Random Forest Tree Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')

X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

```python
y_pred = regressor.predict([[6.5]])
```

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid))
plt.show()
```

```python
regressor100 = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor100.fit(X, y)

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor100.predict(X_grid))
plt.show()
```

```python
regressor300 = RandomForestRegressor(n_estimators = 300, random_state = 0)
regressor300.fit(X, y)

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor300.predict(X_grid))
plt.show()
```
